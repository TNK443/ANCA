###**DESCRIÇÃO - Bundling Newspapers**

```python
# ============================================================
#                     BUNDLING NEWSPAPERS
# ============================================================
# Uma empresa de pesquisa estatística precisa de um programa para produzir instruções para um aplicativo existente que
# cria questionários sobre jornais. O aplicativo requer instruções que são listas de jornais delimitadas por vírgulas,
# cada lista consistindo em um subconjunto do conjunto de jornais em alguma área metropolitana.
#
# Seu programa lerá uma lista de jornais e produzirá essa saída em uma ordem específica e para subconjuntos de alguns tamanhos.
#
# ==========
#   INPUT
# ==========
# A primeira linha da entrada é um inteiro M, depois uma linha em branco seguida por M conjuntos de dados.
# Existe uma linha em branco entre os conjuntos de dados. Cada conjunto de dados consiste em uma primeira linha que descreve os
# tamanhos dos subconjuntos a serem produzidos, seguido pelos nomes dos jornais, um por linha.
#
# Cada linha é encerrada por um marcador de fim de linha e o arquivo de entrada é encerrado por um marcador de fim de arquivo.
# Os nomes dos jornais conterão no máximo 30 caracteres de busca, e haverá no máximo 12 nomes de jornais.
#
# A primeira linha conterá um único asterisco na coluna um, um inteiro positivo ou dois inteiros positivos, o primeiro menor que o segundo.
# Quaisquer números inteiros nesta linha não serão maiores do que o número de jornais a seguir. O significado deles é o seguinte:
#
# (*) significa mostrar todos os tamanhos de subconjunto de 1 ao número de jornais na lista.
#
# (a, b) significa mostrar todos os tamanhos de subconjunto de a até b, inclusive.
#
# (n) significa mostrar apenas subconjuntos de tamanho n.
#
# ==========
# OUTPUT
# ==========
# A saída para cada conjunto de dados deve consistir em subconjuntos de papéis delimitados por vírgulas,
# um subconjunto por linha, agrupado por tamanho de subconjunto e com uma linha de cabeçalho acima
# de cada grupo de tamanho de subconjunto indicando o tamanho.
#
# Imprima uma linha em branco após cada grupo de tamanho de subconjunto.
#
# Dentro de um grupo de tamanho de subconjunto, a ordem das linhas de saída deve ser listada de forma que
# se os nomes do primeiro, segundo e terceiro jornais listados no arquivo de entrada fossem A, B, C, etc., respectivamente,
# o as linhas de saída estariam em ordem lexicográfica (no exemplo acima, A, B; A, C; A, D; B, C; ...).
#
# Imprima uma linha em branco entre os conjuntos de dados.
# 
# ============================================================
```

###**CODIGO**

```python
# ===== CODIGO I ==========
def combs(a):
    if len(a) == 0:
        return [[]]
    cs = []
    for c in combs(a[1:]):
        cs += [c, c+[a[0]]]
    return cs

if __name__ == '__main__':
   try:
    # INPUT
    # The first line of the input is an integer M,... 
    M = int(input())
    # ...then a blank line followed byMdatasets. There is a blankline between datasets.
    input()

    # for m in range(M):
    while (M > 0):
        # Each dataset consists of a first line describing the sizes of subsets to be output,...
        sizesSub = input().split()

        # ...followed by thenames of the newspapers, one per line.
        newspaper=[]
        auxInput=''
        while True:
          # Each line is terminated by an end-of-line marker, and the inputfile is terminated by an end-of-file marker.
          auxInput = input()
          # The newspaper names will contain at most 30 characterseach, and there will be a maximum of 12 newspaper names.
          if (auxInput == '') or (len(newspaper)>=12): break
          newspaper.append(auxInput[:30])
          
        #   (*) means show all subset sizes from 1 to the number of newspapers in the list
        # (a b) means show all subset sizes from a to b, inclusive
        #   (n) means show only subsets of size n

        combinacao = combs(newspaper[::-1])

        if (len(sizesSub)==2):
          a = int(sizesSub[0])
          b = int(sizesSub[1])        
        elif (sizesSub[0]=='*'):
          a = 1
          b = int(len(newspaper))
        else:
          a = int(sizesSub[0])
          b = a
        
        for i in range(a, b+1):
          print(f'Size {i}')
          for k in combinacao[::-1]:
              if len(k) == i: print(', '.join(k))
          print('')     
        
        # Próximo Size (SubConjunto)
        M-=1
        if (M>0):print('')
   except EOFError:
    pass

   finally:
    exit(0)
   exit(0)
```
